```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./times.csv', names=['module', 'size', 'time'], delimiter=';', skipinitialspace=True)
```

```python
data['module'] = data.module.apply(lambda x: x.strip())
```

```python
data = data[data['size'] <= 1000]
```

```python
data2 = data.groupby(['size', 'module']).agg({'time': 'mean'})
```

```python
data3 = data2.unstack('module')
```

```python
#data3.columns = data3.columns.get_level_values(0)
```

```python
data3.columns = data3.columns.get_level_values(1)
data3
```

```python
data3.plot(figsize=(10,10), loglog=True)
#plt.savefig("graph.png")
plt.show()
```

```python
data3[data3.index > 10].plot(figsize=(10,10), loglog=True)
#plt.savefig("graph.png")
plt.show()
```
