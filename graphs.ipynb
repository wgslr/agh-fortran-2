```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./times.csv', names=['module', 'size', 'time'], delimiter=';', skipinitialspace=True)
```

```python
data['module'] = data.module.apply(lambda x: x.strip())
```

```python
data = data[data['size'] <= 1000]
```

```python
data['area'] = data['size']**2


colors = {'naive':'red', 'dot':'blue', 'dotcache':'green', 'builtin':'black', 'cache': 'yellow'}

data['c'] = data['module'].apply(lambda x: colors[x])

# data.groupby('module').plot(x = 'size', y='time',  loglog=True)
data.plot(x = 'size', y='time', loglog=True, kind='scatter', c=data.c, legend=True, figsize=(15,15))
plt.show()
# data.plot(x = 'area', y='time', loglog=True)
# plt.show()
```
